# Data Collection Stage 
Step 1: Create dataframes for each of the seven countries selected for analysis 

```python
# Using the read_csv function and the Path module, to read various country csv files from the data folder.
# Set the `year` as the index column. 

wb_australia = pd.read_csv(
    Path('../data\WB_DATA_AUS.csv'), index_col='year'
    )
wb_brazil = pd.read_csv(
    Path('../data\WB_DATA_BRA.csv'), index_col='year'
    )
wb_china= pd.read_csv(
    Path('../data\WB_DATA_CHN.csv'),  index_col='year' 
    )
wb_germany= pd.read_csv(
    Path('../data\WB_DATA_DEU.csv'), index_col='year' 
    )
wb_nigeria= pd.read_csv(
    Path('../data\WB_DATA_NGA.csv'), index_col='year'
    )
wb_saudi= pd.read_csv(
    Path('../data\WB_DATA_SAU.csv'),  index_col='year' 
    )
wb_usa= pd.read_csv(
    Path('../data\WB_DATA_USA.csv'), index_col='year'
    )
# Review the DataFrame with both the head & tail functions
wb_australia
```

Step 2: Extract the  AG.LND.AGRI.ZS series from the country dataframe for analysis.This series contains percentage of agricultural land area in a country. Agricultural land refers to the share of land area that is arable, under permanent crops, and under permanent pastures.

```python
#  Extracting AG.LND.AGRI.ZS series from the country dataframe for analysis 
#  Create  a new dataframe for the country with 'value' colunm for the percentage of areas land in a country  
Australia_agri_df = wb_australia [wb_australia ['series_id']=='AG.LND.AGRI.ZS'].drop(['country_code','country_name','series_id'], axis=1)

Australia_agri_df.rename(columns={'value':'Australia'},inplace=True)
Australia_agri_df.sort_index
Australia_agri_df.tail()
```

```python
#  Extracting AG.LND.AGRI.ZS series from the country dataframe for analysis 
#  Create  a new dataframe for the country with 'value' colunm for the percentage of areas land in a country  

Brazil_agri_df = wb_brazil [wb_brazil ['series_id']=='AG.LND.AGRI.ZS'].drop(['country_code','country_name','series_id'], axis=1)

Brazil_agri_df.rename(columns={'value':'Brazil'},inplace=True)

Brazil_agri_df.tail()
```

Using the 5-year rolling window, the agric land area data for  China and Brazil s, calculate the covariance of  Agric Land Area allocation for the two countries.

```python
china_brazil_rolling_agric_convarince= wb_countries_agri['China'].rolling(window=5).cov( wb_countries_agri['Brazil']).dropna()

china_brazil_rolling_agric_convarince.head(20)
```
