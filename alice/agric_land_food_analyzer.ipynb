# Further Analysis of China and Brazil  Data 
## Using the 5-year rolling window, the agric land area data for  China and Brazil s, calculate the covariance of  Agric Land Area allocation for the two countries.

```python
#Create a new dataframe of the convariance between  China and Brazil's  agric land area and food productiion using a 5 year rolling window. 
china_brazil_rolling_agric_convarince= wb_countries_agri['China'].rolling(window=5).cov( wb_countries_agri['Brazil']).dropna()

china_brazil_rolling_agric_convarince.tail()
```

```python
#Visualize the convarinance for China  and Brazil  agric land area
china_brazil_rolling_agric_convarince.hvplot.line(label="China_Brazil_Agric_Land Area Convariance",color="blue",frame_width =800,
    frame_height = 400) 
```

There is a strong positive relationship of the agriculrure land area  for China and Brazil in the perriod 1965 to 1986 meaning that they tended to similarly move up and down on a yearly basis. After 1986 this relationship is weak. 
 
This may imply either country may have adopted different policy direction on agriacultural land area in the latter years of the analyisis.
