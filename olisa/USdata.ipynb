```python
import pandas as pd
from pathlib import Path 
%matplotlib inline
```

```python
worldbank_us_df = pd.read_csv(Path("../data/WB_DATA_USA.csv"))
```

```python
worldbank_us_df.head()
```

```python
worldbank_us_co2_df = worldbank_us_df[worldbank_us_df["series_id"]=="EN.ATM.CO2E.KT"]
```

```python
worldbank_us_co2_df.head(20)
```

```python
worldbank_us_agr_df = worldbank_us_df[worldbank_us_df["series_id"]=="AG.LND.AGRI.ZS"]
```

```python
worldbank_us_agr_df.head(20)
```

```python
worldbank_us_co2_df.set_index("year",inplace=True)
```

```python
worldbank_us_co2_df.head()
```

```python
worldbank_us_co2_df.isnull().sum()
```

```python
worldbank_us_co2_df.drop_duplicates(inplace=True)
```

```python
worldbank_us_co2_df.loc[1990:2000]["value"].head()
```

```python
worldbank_us_co2_df.loc[1990:2010]["value"].plot(kind='line',title='My graph',xlabel='Year',ylabel='CO2 Emissions in Kilotons (Millions)')
```

# Analysis
* CO2 emissions steadily increased from 1991 to 2000. It maintained a steady rate of 5.6-5.8Kt from 2000 to mid 2007 and dropped drastically in 2008
